# Complex Numbers

The number system began with natural numbers for counting, later extended (adjoining the additive inverse) to integers to include negative numbers. Then By allowing quotients (fractions) of integers to rational numbers (form a field). Eventually to real numbers to capture irrational numbers like $\sqrt(2), \pi$.

Yet some equations, such as $x^2+1=0$ had no solution within the real numbers. To address this mathematicians introduced a new unit i, defined by $i^2 = -1$. Numbers of form $a+bi$ where a and b are real numbers became the complex numbers. 

The hierarchy
$\mathbb{N}\subset\mathbb{Z}\subset\mathbb{Q}\subset\mathbb{R}\subset\mathbb{C}$ illustrates successive algebraic closures of the number systems.


Rational numbers are written as an ordered pairs of integer (n,m). They follow simple rules:
- (n,m) + (p,q) = (nq+pm, mq)
- (n,m)(p,q) = (np, mq)

Similar rules can be applied for complex numbers $a + bi$ as (a,b)
- (a,b) + (c,d) = (a+c, b+d)
- (a,b)(c,d) = (ac-bd, ad+bc)

Subtraction and division are the inverse of these operations.
